```python
phylogeny_path = '/home/harishbabu/data/phlyogenyCUB/1_tree-consensus-Hacket-27Species-modified_cub-names_v1.phy'
root = construct_phylo_tree(phylogeny_path)
root.assign_all_descendents()
```

```python
def decendent_leafs(node):
    num_decendent_leafs = 0
    bfs = [node]
    while len(bfs) > 0:
        new_bfs = []
        for child in bfs:
            if child.num_children() > 1:
                new_bfs += child.children
            else:
                num_decendent_leafs += 1
        bfs = new_bfs

    return num_decendent_leafs


# node_desc = [(node, len(node.descendents)) for node in root.nodes_with_children()]

node_desc = [(node, decendent_leafs(node)) for node in root.nodes_with_children()]
node_desc = sorted(node_desc, key=lambda x: x[1])[::-1]

for node, num_descendents in node_desc:
    print(node.name, num_descendents)
```
